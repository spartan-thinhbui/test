### Step 1: Use the model via the API

```python
# test the server by running client.py or this code
import requests

response = requests.post("http://127.0.0.1:8000/predict", json={"input": 4.0})
print(f"Status: {response.status_code}\nResponse:\n {response.text}")
```

### Step 2: Expose the server to the internet
Use the API builder Studio plugin on the right to allow anyone on the internet to use this server



### Step 3: Enable auto start (serverless)
Auto start (aka serverless) turns off the Studio when the model is not in use. When it becomes active, it turns back on to serve the requests it receives.  

Auto start requires knowing what command to run when the Studio auto-starts. This has already been done for you in the endpoint that's been created for you
